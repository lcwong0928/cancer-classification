```python
import pickle
import numpy as np
from sklearn.decomposition import SparsePCA
import matplotlib.pyplot as plt
```

```python
with open('data/combined_data_breast', 'rb') as f:
    X = pickle.load(f)
with open('data/combined_y_breast', 'rb') as f:
    y = pickle.load(f)
```

```python
transformer = SparsePCA(n_components=5, normalize_components=False, random_state=0)
transformer.fit(X) 
```

```python
X_transformed = transformer.transform(X)
X_transformed.shape
```

```python
plt.plot(X_transformed[:, 0][y==0], X_transformed[:, 1][y==0], '.')
plt.plot(X_transformed[:, 0][y==1], X_transformed[:, 1][y==1], '.')
```

```python
print(X.shape)
plt.plot(X_transformed[:, 0], '.')
```

```python
pickle_out = open("sparse.pickle","wb")
pickle.dump(transformer, pickle_out)
pickle_out.close()
```
